```python
city = ['Seattle','San Francisco','Los Angeles','Denver','Salt Lake City','Phoenix','Chicago','St. Louis','Dallas','Boston','Washington D.C.','Miami']
Poorday_Rate20=[16,25,157,33,20,69,27,9,28,0,2,2]
Poorday_Rate10=[3,14,115,38,24,58,43,103,28,10,50,6]
```

```python
import matplotlib.pyplot as plt
import numpy as np
alldata = np.arange(len(city))

plt.style.use('fivethirtyeight')
plt.bar(alldata+.5,Poorday_Rate10, color='blue', label='2010', width=.3)
plt.bar(alldata,Poorday_Rate20, color='red', label='2020', width=.3)
plt.title('Rate of Poor Days by City in 2010 and 2020')
plt.legend()
plt.xlabel('City')
plt.ylabel('Poor Day Rate')
plt.grid(True)
plt.xticks(ticks = alldata, labels =city)
plt.show()
```
